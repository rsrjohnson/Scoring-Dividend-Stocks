```python
import numpy as np
import yahoo_fin.stock_info as si
import re
import pandas as pd


stocks = pd.read_csv('stocks.csv')

for index,stock in enumerate(stocks["Ticker"]):
    
    try:
        q = si.get_quote_table(stock.lower())
        
        df = pd.DataFrame(list(q.values()),index=list(q.keys())).T
        
        if index == 0:
            df_f = df
        else:
            df_f = pd.concat([df_f, df])



    except:
        print(stock)
```

```python
display(df_f.dropna().sort_values(by=["Forward Dividend & Yield"],ascending=False))
```
